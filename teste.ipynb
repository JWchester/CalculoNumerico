```python
from sympy import *
import math
def f(x):
    return  2*cos(3*x)-3*sin(2*x)


n = 2 # grau do polinomio
x = [3,3.5,4] # valores de x a serem usados
fx = [] # y da função
_x = Symbol('x')
p = [] # são os polinomios

#Cálculo das imagens nos pontos
print( 'Resolvendo pela forma de lagrange temos:')
print()
print()
print('(simbolo do somatorio e produtorio)')
print()

print("calculando a fx que irá compor o resultado na forma de lagrange: ")
print()
for i in range(n+1):
    fx.append(round(f(x[i]),4))
    print("f({}) = 2*cos(3*{})-3*sin(2*{}) = {} ".format(x[i],x[i],x[i],fx[i]))
print()    

print('calculando o produtório para cada ponto:')

#Produtório
for i in range(n+1):
    _p = fx[i]
    print('p({}) = {}*'.format(x[i],fx[i]), end='')
    for j in range(n+1):
        if i != j:
            _p *= ((_x - x[j]) / (x[i] - x[j]))
            print("(({} - {}) / ({} - {})) * ".format('x',x[j],x[i],x[j]),end= ' ')
    print('= ',end="")
    print(_p)
    p.append(_p.expand())
    print()
    print('resultado do polinômio p({}) = {})'.format(i,p[i]))
    print()
print()
print()

#Somatório
_sum = 0
print()
print("por fim temos o resultado da soma dos polinômios que irá compor a função: ")
for i in p:
    _sum += i
print(_sum)
```
